```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result.csv')
```

```python
sorted_df = df.sort_values(by='Accuracy', ascending=False)
sorted_df
```

```python
# Accuracy
sns.barplot(x='Accuracy', y='Model', data=df,palette='viridis')
```

```python
# confusion matrix
matrix = np.load('confusion_matrix.npy')
for i in range(0, len(matrix)):
    plt.figure(figsize=(10,7))
    sns.heatmap(matrix[i], annot=True, cmap='viridis',fmt='.5f')
    plt.title(df['Model'][i])
    plt.show()
```
